```python
import matplotlib.pyplot as plt

# x-axis (only 5..25)
k_vals = [5, 10, 15, 20, 25]

# your ASR values
asr_white_box = [0.97, 0.89, 0.88, 0.80, 0.69]

plt.figure()
plt.plot(k_vals, asr_white_box, marker='o', label="White-Box")

# axes formatting
plt.ylim(0.0, 1.0)            # y-axis from 0 to 1
plt.xlim(5, 25)               # x-axis span
plt.xticks(k_vals)            # only show these ticks

plt.xlabel("k")
plt.ylabel("ASR")
plt.title("ASR vs k")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
